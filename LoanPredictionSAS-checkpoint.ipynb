# Creating macro for Proc Freq and Proc Sgplot functions

```sas
%macro sgplotchart(var,ds);
title " &ds Dataset";
proc sgplot data = &ds;
vbar &var / stat = sum;
xaxis display = all;
run;
%mend sgplotchart;
```

```sas
%macro freqtable (var,ds);
 title "&ds";
 proc freq data = &ds;
 tables &var / nocum;
 run;
%mend freqtable;

```

```sas
title;
%freqtable(Married,lp)
%freqtable(Married,lptest)
```

# Tidying the data (Imputing missing values)

```sas
proc print data = lp (obs =5);
run;
proc print data = lptest(obs =5);
run;
```

 Combining the training and test datasets

```sas
data lpcombine;
set lp lptest;
run;
```

```sas
proc print data = lpcombine (firstobs = 970 obs=981);
run;
```

```sas
data lpcombine;
    set lpcombine;
    
if Married = " " and CoApplicantIncome = 0
    then Married = "No";
else if Married = " " then
    Married = "Yes";
run;
* Single with Coapplicantincome =0 can be termed as unmarried.
```

```sas
proc freq data = lpcombine;
tables Married/ nocum;
run;
```

Imputing Gender and dependents

```sas
data lpcombine;
    set lpcombine;
    if Gender =" " and Dependents = " "
        then Gender = "Male";
run;
```

```sas
%crosstab(Gender,Dependents,lpcombine)
```

Since the mode for No. of Dependents for Male and Female is 0, we can impute the same...

```sas
data lpcombine;
    set lpcombine;
    if Dependents = " " and Married = "No"
        then Dependents = "0";
run;
```

```sas
data lpcombine;
    set lpcombine;
    if Gender = "Male" and Married = "Yes"
    then Dependents = "0";
run;
```

```sas
%freqtable(Dependents, lpcombine)
```

```sas
data lpcombine;
     set lpcombine;
     if Gender = "M" then Gender = "Male";
run;
```

```sas
%freqtable(Gender, lpcombine)
```

Imputing LoanAmount 

```sas
proc mi data=lpcombine seed=1213 nimpute=2 mu0=50 10 180 out=outex6;
   fcs nbiter=10 reg(/details);
   var LoanAmount Loan_Amount_Term;
run;
```
